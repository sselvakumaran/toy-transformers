```python
import sys
import os
import torch
from pathlib import Path

def get_project_info() -> Path:
  current = Path.cwd().resolve()
  root = current
  for parent in [current, *current.parents]:
    if (parent / "toy_transformers").exists():
      root = parent
      break
  return root, current

if 'ROOT_DIR' not in globals():
	ROOT_DIR, EXPERIMENT_DIR = get_project_info()
	if str(ROOT_DIR) not in sys.path:
		sys.path.append(ROOT_DIR)
	if Path.cwd() != ROOT_DIR:
		os.chdir(ROOT_DIR)
```

```python
from toy_transformers.pipelines import tokenization
tokenization.tokenize_from_scratch(
  "data/input.txt",
  EXPERIMENT_DIR / "artifacts/vocab256",
  EXPERIMENT_DIR / "artifacts/data",
  256,
  verbose=True
)
```
