```python
import sys
import os
from tqdm import tqdm

sys.path.append(os.path.abspath(os.path.join('..')))

from utilities.loss_plot import LossTracker
```

```python
import time
import random
from IPython.display import display, clear_output
plotter = LossTracker(palette="deep")
```

```python
print("running")
for i in range(100):
  train_loss = 1 / (i + 1) + random.uniform(0, 0.05)
  val_loss = 1 / (i + 2) + random.uniform(0, 0.05) if i % 10 == 0 else None
  plotter.add_points(train_loss, val_loss)
  fig = plotter.render_fast_figure()
  clear_output(wait=True)
  display(fig)
  time.sleep(0.05)
```
